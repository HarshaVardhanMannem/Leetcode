```python
from typing import List

class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        def search(big, small):
            big.sort()                 # required for binary search
            small = sorted(set(small)) # deduplicate + sort
            res = []
            for num in small:
                low, high = 0, len(big) - 1
                while low <= high:
                    mid = (low + high) // 2
                    if big[mid] == num:
                        res.append(num)
                        break
                    elif big[mid] > num:
                        high = mid - 1
                    else:
                        low = mid + 1
            return res
        if len(nums1) > len(nums2):
            return search(nums1, nums2)
        else:
            return search(nums2, nums1)

test = Solution()
print(test.intersection([1,2,2,1],[2,2]))



```

```python
class Solution:
    def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
        res=[]
        if len(nums1)>len(nums2):
            for num in nums2:
                if num in nums1:
                    res.append(num)
                    nums1.remove(num)
        else:
            for num in nums1:
                if num in nums2:
                    res.append(num)
                    nums2.remove(num)
        return res
        
test = Solution()
print(test.intersect([1,2,2,1],[2,2]))
 
```

## Intersection of Two Arrays

### Problem Explanation
Find the intersection of two arrays (unique common elements).

---

### Solution 1: Set Intersection (Simple Approach)
**Time Complexity:** O(n + m) where n and m are lengths of arrays  
**Space Complexity:** O(n + m)

**Explanation:**
- Convert both arrays to sets to remove duplicates
- Use set intersection operator `&` to find common elements
- Convert back to list
- **Pros:** Simple, readable, efficient for small arrays
- **Cons:** Uses extra space for sets

---

### Solution 2: Binary Search Approach
**Time Complexity:** O(n log n + m log m) for sorting, O(m log n) for searching  
**Space Complexity:** O(min(n, m)) for result

**Explanation:**
- Sort the larger array (required for binary search)
- Deduplicate and sort the smaller array
- For each element in the smaller array, use binary search to find it in the larger array
- **Key Steps:**
  1. Compare array lengths to determine which to sort (optimize by sorting the larger one)
  2. Binary search: repeatedly divide search space in half
  3. If found, add to result and break
- **Pros:** More space-efficient, good when one array is much larger
- **Cons:** Requires sorting, more complex implementation

---

### Solution 3: Intersection with Duplicates (Intersect II)
**Time Complexity:** O(n Ã— m) in worst case  
**Space Complexity:** O(min(n, m))

**Explanation:**
- Iterate through the smaller array
- For each element, check if it exists in the larger array
- If found, add to result and remove from larger array (to handle duplicates correctly)
- **Note:** This solves a different problem - preserves duplicate counts
- **Pros:** Handles duplicates, simple logic
- **Cons:** Inefficient due to `remove()` operation (O(n) each time)
